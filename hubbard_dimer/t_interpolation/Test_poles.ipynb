```python
import numpy as np
import matplotlib.pyplot as plt

# Real-valued z
z = np.linspace(0, 10, 1000)

# Function 
f = (1-2*1j) / (z - 5 - 0.45*1j)  + 1 / (z - 1 - 0.1*1j) - 7*1j / (z - 8 - 1j) - 7*1j / (z - 1.5 - 0.1*1j)
# Real and imaginary parts
f_real = np.real(f)
f_imag = np.imag(f)

# Plot
plt.figure(figsize=(8,5))

plt.plot(z, f_real, label='Re[f(z)]')
plt.plot(z, f_imag, label='Im[f(z)]')

plt.xlabel('z')
plt.ylabel('Function value')
plt.title(r'Real and Imaginary Parts of $f(z)=1/(z-2-i)$')
plt.legend()
plt.grid(True)
plt.show()
```
