```python
def G0_bare(N_sites, wn, U, t=1.0):
    """
    Compute the bare Green's function G^{(0)}_{ij}(iω_n) for the 1D Hubbard model
    at half-filling, with chemical potential μ = U/2 (MBPT convention).

    Parameters:
        N_sites : int
            Number of lattice sites
        beta : float
            Inverse temperature β
        U : float
            Hubbard U parameter
        t : float
            Hopping parameter (default=1.0)

    Returns:
        G0 : np.ndarray
            Bare Green's function of shape (N_omega, N_sites, N_sites)
    """
    G0 = np.zeros((wn.size, N_sites, N_sites), dtype=np.complex128)
    k_vals = 2 * np.pi * np.arange(N_sites) / N_sites

    # MBPT dispersion relation: ε_k = -2t cos(k) 
    eps_k = -2 * t * np.cos(k_vals) 

    for n in range(wn.size):
        Gk = 1.0 / (1j * wn[n] - eps_k)  # G^{(0)}(k, iω_n)
        for i in range(N_sites):
            for j in range(N_sites):
                phase = np.exp(1j * k_vals * (i - j))
                G0[n, i, j] = np.sum(Gk * phase) / N_sites

    return G0
```
